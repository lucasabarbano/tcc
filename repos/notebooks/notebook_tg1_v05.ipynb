# Imports

```python
import seaborn         as sns
import pandas          as pd
from statsmodels.tsa.seasonal import seasonal_decompose


from matplotlib              import pyplot as plt
from IPython.core.display    import HTML
```

## Helper Functions

```python
def jupyter_settings():
    %matplotlib inline
    %pylab inline
    
    plt.style.use( 'bmh' )
    plt.rcParams['figure.figsize'] = [25, 12]
    
    display( HTML( '<style>.container { width:100% !important; }</style>') )
    pd.options.display.max_columns = None
    pd.options.display.max_rows = None
    pd.set_option( 'display.expand_frame_repr', False )
    
    sns.set()
    
jupyter_settings()
```

## Loading data

```python
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m')
data = pd.read_csv('../data/AirPassengers.csv', parse_dates=['Month'], index_col='Month',date_parser=dateparse)
```

```python
data.head()
```

```python
data.tail()
```

# Times Series Visualisation

```python
plt.plot( data )
plt.xlabel('Anos', fontsize = 24)
plt.ylabel('Passageiros', fontsize = 24)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```

# Time Series Decomposition

```python
result = seasonal_decompose( data, model='multiplicative' )
result.plot()
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```

```python
result2 = seasonal_decompose( data, model='aditive' )

result2.plot(observed=False)
plt.xlabel('Anos', fontsize = 24)
plt.ylabel('Sazonalidade', fontsize = 24)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```
